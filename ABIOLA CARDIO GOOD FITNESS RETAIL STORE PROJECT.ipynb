# Analyses for Customers Education

```python
#Using seaburn and matplotlib with countplot for Visualisation

#This uses seaburn countplot function to create a bar plot showing the count of each category in the result ploted in variable.
ax = sns.countplot(data=df, x='Education')

#calculate the total numbers of observation in the "Education" column and store it in the variable "total" as floating number.
total = float(len(df['Education']))

#The following loop iterate through each bar in the countplot
for p in ax.patches:

#This calculate the percentage of each category in the count relative to the total number of observations.    
    percentage = '{:.1f}%'.format(100 * p.get_height()/total)
    x = p.get_x() + p.get_width() / 2.
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center', va='center', xytext=(0, 10), textcoords='offset points')

#Using matplotlib function to set the title of the plot to countplot with percentages
plt.title('Countplot with Percentages')

#this display the plot that has been created with seaborn and annotated with percentage
plt.show()
```

```python
#visuliasation of Education distribution 
#Purpose is to basically show the description of variable type, which is a numeric discrete variable.
#Showing the numbers of years a Customer has been in aware about a product
```
